```python
import cv2
import matplotlib.pyplot as plt

img =cv2.imread("/content/mj.jpg",0);

#Apply Histogrsm Equalization
he_img=cv2.equalizeHist(img)

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.title("Orginal Image")
plt.imshow(img,cmap="gray")
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Histogram Equalization")
plt.imshow(he_img,cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram (Equalized)")
plt.hist(img.ravel(),256)
plt.show()
```

```python
#Point Processing
#color Enhancement
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("/content/girl.avif")
img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
# Split Y, Cr, Cb
Y, Cr, Cb = cv2.split(img_ycrcb)
# CLAHE on Y channel
clahe = cv2.createCLAHE(clipLimit=2.5, tileGridSize=(8,8))
Y_clahe = clahe.apply(Y)
# Merge & convert back
img_clahe = cv2.merge((Y_clahe, Cr, Cb))
img_clahe = cv2.cvtColor(img_clahe, cv2.COLOR_YCrCb2BGR)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Color CLAHE")
plt.imshow(cv2.cvtColor(img_clahe, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

#loading three different imager
#this methods for smooth tech

img_avg = cv2.imread('/content/leafs.webp')
img_gaussian = cv2.imread('/content/ed.jpeg')
img_median = cv2.imread('/content/charli.png')

#showing images
#checking loading
if img_avg is None or img_gaussian is None or img_median is None:
    print('Error loading image')
    exit()

#covert BGR -> RBG for correct display
img_avg_rgb = cv2.cvtColor(img_avg, cv2.COLOR_BGR2RGB)
img_gaussion_rgb = cv2.cvtColor(img_gaussian, cv2.COLOR_BGR2RGB)
img_median_rgb = cv2.cvtColor(img_median, cv2.COLOR_BGR2RGB)

#applying filters

#1.Average Soothing
avg_blur = cv2.blur(img_avg_rgb, (5, 5))
avg_blur_rgb = cv2.cvtColor(avg_blur, cv2.COLOR_BGR2RGB)
#2.Gaussian Soothing
gaussian_Blur = cv2.GaussianBlur(img_gaussion_rgb, (5, 5), 0)
gaussian_Blur_rgb = cv2.cvtColor(gaussian_Blur, cv2.COLOR_BGR2RGB)
#3.median Soothing
median_blur = cv2.medianBlur(img_median_rgb, 5)
median_blur_rgb = cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)


#Displaying befour and after for each filter

plt.figure(figsize=(14, 10))
#avg
plt.subplot(3, 2, 1)
plt.title('Original Image(Average Blur)')
plt.imshow(img_avg_rgb)
plt.axis('off')

plt.subplot(3, 2, 2)
plt.title('After Average Blur')
plt.imshow(avg_blur_rgb)
plt.axis('off')

#gaussian
plt.subplot(3, 2, 3)
plt.title('Original Image(Gaussian Blur)')
plt.imshow(img_gaussion_rgb)
plt.axis('off')

plt.subplot(3, 2, 4)
plt.title('After Gaussian Blur')
plt.imshow(gaussian_Blur_rgb)
plt.axis('off')

#median
plt.subplot(3, 2, 5)
plt.title('Original Image(Median Blur)')
plt.imshow(img_median_rgb)
plt.axis('off')

plt.subplot(3, 2, 6)
plt.title('After Median Blur')
plt.imshow(median_blur_rgb)
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img=cv2.imread("/content/ntr.jpeg",0);

#sharpening kernel (simple 3x3 filter)

kernel =np.array([[0, -1, 0],
                 [-1, 5, -1],
                 [0, -1, 0]])

#appling the sharpening
sharpened = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('on')

plt.subplot(1, 2, 2)
plt.title('Sharpened Image')
plt.imshow(sharpened, cmap='Blues')
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img=cv2.imread("/content/ntr.jpeg",0);

#1.sobel edge
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)#vertical edges
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)#horizontal edges
sobelxy = cv2.magnitude(sobelx, sobely)

#2,canny edge detection
canny_edges = cv2.Canny(img, 100, 200)

#display
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel Edge Detection')
plt.imshow(sobelxy, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Canny Edge Detection')
plt.imshow(canny_edges, cmap='gray')
plt.axis('off')

plt.show()


```
